```python
def normalize_name(row):
    if pd.isna(row['Name']):
        return row['Name'] 

    parts = row['Name'].split()
    
    if len(parts) > 1:
        first_name = parts[0]
        last_name = parts[-1].capitalize()
        return f"{first_name} {last_name}"
    else:
        return row['Name']
```

```python
combined_df['Name'] = combined_df.apply(normalize_name, axis=1)
```

```python
def normalize_name_last_first(row):
    if pd.isna(row['Name']):
        return row['Name'] 
    
    if row['Year'] in [2020, 2024]:
        parts = row['Name'].split()
        if len(parts) > 1 and parts[0].isupper():  
            first_name = parts[1].capitalize()     
            last_name = parts[0].capitalize()      
            return f"{first_name} {last_name}"
    
    return row['Name']
```

```python
df= combined_df.copy()
df['Name'] = combined_df.apply(normalize_name_last_first, axis=1)
```
